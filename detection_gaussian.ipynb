### A. Détecteur Bayésien

#### 1. Expression des densités de probabilités

On observe un vecteur aléatoire $\boldsymbol{Y}=[Y[1], ..., Y[N]]^{\mathsf{T}}$ dont les composantes s'écrivent:
$$
\begin{aligned}
    \forall n \in [1, N],\ 
    \begin{cases}
        \mathbb{H}_0: Y[n] &= m_0 + V[n], m_0\in\mathbb{R}\\
        \mathbb{H}_1: Y[n] &= m_1 + V[n], m_1\in\mathbb{R}
    \end{cases}
\end{aligned}
$$
où les termes de bruit forment un vecteur $\boldsymbol{V}=[V[1], ..., V[N]]^{\mathsf{T}}$.

On a supposé dans l'énoncé que les composantes de $\boldsymbol{V}$ sont **identiquement et indépendamment distribuées** (iid) selon
$$
\begin{aligned}
    \forall n \in [1, N],\ V[n] \sim \mathcal{N}(0, \sigma^2).
\end{aligned}
$$
ce qui implique que les composantes de $\boldsymbol{Y}$ sont elles aussi **iid** selon
$$
\begin{aligned}
    \forall n \in [1, N],\
    \begin{cases}
        \mathbb{H}_0: Y[n] \sim \mathcal{N}(m_0, \sigma^2)\\
        \mathbb{H}_1: Y[n] \sim \mathcal{N}(m_1, \sigma^2)
    \end{cases}.
\end{aligned}
$$
On peut donc écrire que
$$
\begin{aligned}
    \forall i \in \{0,1\},\ 
        f_{\boldsymbol{Y}\mid M}(\boldsymbol{y}\mid m_i)
        &=
            \prod_{n=1}^{N}
                f_{Y[n]\mid M}(y[n]\mid m_i)
                &\text{(indépendance)}
        \\
        &=
            \prod_{n=1}^{N}
                \frac{1}{\sqrt{2 \pi \sigma^2}}
                \exp\left(
                    -\frac{
                        (y[n] - m_i)^{2}
                    }{
                        2\sigma^2
                    }
                \right)
                &\text{(distribution identique).}
        \\        
\end{aligned}
$$

Pour aller un peu plus loin dans les calculs, on peut remarquer que
$$
\begin{aligned}
    \forall i \in \{0,1\},\ 
        f_{\boldsymbol{Y}\mid M}(\boldsymbol{y}\mid m_i)
        &=
            \left(\frac{1}{\sqrt{2 \pi \sigma^2}}\right)^N
            \exp\left(
                -\sum_{n=1}^{N}
                    \frac{
                        (y[n] - m_i)^{2}
                    }{
                        2\sigma^2
                    }
            \right)
                &\text{(produit d'exp. }\rightarrow\text{ exp. de somme)}
        \\
        &=  
            \left(\frac{1}{\sqrt{2 \pi \sigma^2}}\right)^N
            \exp\left(
                -\frac{
                    \|\boldsymbol{y} - m_i\boldsymbol{1}_{N,1}\|_2^{2}
                }{
                    2\sigma^2
                }
            \right)
                &\text{(norme euclidienne au carré).}
\end{aligned}
$$

#### 2. Rapport de vraisemblance

Le rapport de vraisemblance s'écrit de la façon suivante
$$
\begin{aligned}
    \Lambda(\boldsymbol{y})
    &=
        \frac{
            f_{\boldsymbol{Y}\mid M}(\boldsymbol{y}\mid m_1)
        }{
            f_{\boldsymbol{Y}\mid M}(\boldsymbol{y}\mid m_0)
        }
\end{aligned}
$$
On remplace avec l'expression calculée dans la question d'avant
$$
\begin{aligned}
    \Lambda(\boldsymbol{y})
    &=
        \frac{
            \left(\frac{1}{\sqrt{2 \pi \sigma^2}}\right)^N
            \exp\left(
                -\frac{
                    \|\boldsymbol{y} - m_1\boldsymbol{1}_{N,1}\|_2^{2}
                }{
                    2\sigma^2
                }
            \right)
        }{
            \left(\frac{1}{\sqrt{2 \pi \sigma^2}}\right)^N
            \exp\left(
                -\frac{
                    \|\boldsymbol{y} - m_0\boldsymbol{1}_{N,1}\|_2^{2}
                }{
                    2\sigma^2
                }
            \right)
        }
    \\
    &=  
        \exp\left(
            -\frac{
                \|\boldsymbol{y} - m_1\boldsymbol{1}_{N,1}\|_2^{2}
            }{
                2\sigma^2
            }
            +\frac{
                \|\boldsymbol{y} - m_0\boldsymbol{1}_{N,1}\|_2^{2}
            }{
                2\sigma^2
            }
        \right)    
    \\
    &=  
        \exp\left(
            \frac{
                \|\boldsymbol{y} - m_0\boldsymbol{1}_{N,1}\|_2^{2} - \|\boldsymbol{y} - m_1\boldsymbol{1}_{N,1}\|_2^{2}
            }{
                2\sigma^2
            }
        \right)
    \\
    &=  
        \exp\left(
            \frac{
                \|\boldsymbol{y}\|^2 + Nm_0^2 - 2 m_0 \sum_{n=1}^{N} y[n]
                - \|\boldsymbol{y}\|^2 - Nm_1^2 + 2 m_1 \sum_{n=1}^{N} y[n]
            }{
                2\sigma^2
            }
        \right)
    \\
    &=  
        \exp\left(
            \frac{
                N(m_0^2 - m_1^2) + 2 (m_1 - m_0) \sum_{n=1}^{N} y[n]
            }{
                2\sigma^2
            }
        \right)
\end{aligned}
$$

#### 3. Décision optimale

Pour prendre une décision sur la valeur de $M$, on va faire un test de Bayes optimal. En utilisant le résultat de la question précédente, le risque de Bayes s'écrit:
$$
\begin{aligned}
    \Lambda(\boldsymbol{y})
    \ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
    \eta
    \\
    \exp\left(
        \frac{
            N (m_0^2 - m_1^2) + 2 (m_1 - m_0) \sum_{n=1}^{N} y[n]
        }{
            2\sigma^2
        }
    \right) 
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
    \eta.
\end{aligned}
$$
où
$$
\begin{aligned}
    \eta 
    &= 
        \frac{
            C_{\text{fp}}
        }{
            C_{\text{nd}}
        }
        \frac{
            \mathbb{P}_M(m_0)
        }{
            \mathbb{P}_M(m_1)
        }
\end{aligned}
$$
En appliquant la fonction $\ln$ de chaque côté, on obtient
$$
\begin{aligned}
    \frac{
        N (m_0^2 - m_1^2) + 2 (m_1 - m_0) \sum_{n=1}^{N} y[n]
    }{
        2\sigma^2
    }
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
    \ln \eta.
    \\
    N (m_0^2 - m_1^2) + 2 (m_1 - m_0) \sum_{n=1}^{N} y[n]    
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
    2 \sigma^2 \ln \eta.
    \\
    2 (m_1 - m_0) \sum_{n=1}^{N} y[n]    
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
    2 \sigma^2 \ln \eta + N (m_1^2 - m_0^2)
    \\
    \sum_{n=1}^{N} y[n]    
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\
    \frac{\sigma^2 \ln \eta}{m_1 - m_0} + N\frac{m_1 + m_0}{2}
    \\
    \underbrace{\frac{1}{N} \sum_{n=1}^{N} y[n]}_{S}
    \ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\
    \underbrace{\frac{\sigma^2 \ln \eta}{N(m_1 - m_0)} + \frac{m_1 + m_0}{2}}_{\gamma}
\end{aligned}
$$

#### 4. Détecteur ML et MAP

Pour le ML comme pour le MAP, on n'a pas de notion de coût. Cela revient à supposer que
$$
\begin{aligned}
    C_{fd} = C_{nd} \neq 0.
\end{aligned}
$$

##### Maximum de vraisemblance - ML
Pour une détection ML, on n'a pas de probabilité a priori sur les valeurs que peut prendre $M$
$$
\begin{aligned}
    \mathbb{P}_M(m_0) = \mathbb{P}_M(m_1) &= \frac{1}{2}. &\text{(équiprobabilité)}
\end{aligned}
$$
ce qui conduit à $\eta = 1$. La détection ML consiste donc à effectuer le test suivant
$$
\begin{aligned}
    S
    \ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
    \frac{m_1 + m_0}{2}
\end{aligned}
$$

##### Maximum A Posteriori - MAP
Pour une détection MAP, on suppose qu'on a une information a priori sur les probabilités d'obtenir $M=m_0$ et $M=m_1$
$$
\begin{aligned}
    \mathbb{P}_M(m_0) \neq \mathbb{P}_M(m_1).
\end{aligned}
$$
On a donc $\eta\neq 1$ et le test s'écrit
$$
\begin{aligned}
    S
    \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\
    \frac{\sigma^2}{N}
    \frac{\ln \mathbb{P}_M(m_0) - \ln \mathbb{P}_M(m_1)}{m_1 - m_0} + \frac{m_1 + m_0}{2}
\end{aligned}
$$

#### (Bonus) Simulation

```python
%matplotlib inline
import numpy as np
import ipywidgets as wg
from scipy import stats
from cna import detection
from matplotlib import pyplot as plt

#Main
def main(N, p_0, m_0, m_1, c_fa, c_md, sigma):
    # Prior probabilities for both hypothesis
    p = np.array([p_0, 1 - p_0])

    # Noise mean for both hypothesis
    m = np.array([m_0, m_1])

    # Generate samples of Y
    hypothesis_true = detection.select_hypothesis(p)
    y_samples = detection.generate_samples_gaussian(mean=m[hypothesis_true], std=sigma, n_samples=N)

    # Test
    eta = detection.bayes_threshold(p, c_fa, c_md)
    gamma = detection.test_threshold_gaussian(eta, m, sigma, N)
    hypothesis_test = detection.bayes_test(y_samples, gamma)

    # Figure
    fig, ax = plt.subplots(figsize=[8,3])
    ax.set(
        xlim=[-10, 10],
        ylim=[-0.05, 1 / np.sqrt(2*np.pi*sigma**2) + 0.05],
    )
    
    # axes 
    y = np.linspace(-100, 100, 10000)
    pdf_0 = stats.norm(loc=m[0], scale=sigma).pdf(y)
    pdf_1 = stats.norm(loc=m[1], scale=sigma).pdf(y)
    
    # Plot means
    color_hypothesis = ['b', 'g']
    ax.vlines(m[0], ymin=0, ymax=1/np.sqrt(2*np.pi*sigma**2), lw=1, ls='--', color=color_hypothesis[0], alpha=0.5)
    ax.vlines(m[1], ymin=0, ymax=1/np.sqrt(2*np.pi*sigma**2), lw=1, ls='--', color=color_hypothesis[1], alpha=0.5)
    
    # Plot pdfs
    ax.plot(y, pdf_0, color=color_hypothesis[0], lw=1, label=r'$\mathcal{N}(m_0, \sigma^2)$')
    ax.plot(y, pdf_1, color=color_hypothesis[1], lw=1, label=r'$\mathcal{N}(m_1, \sigma^2)$')
    
    # Plot error probabilities
    ax.fill_between(y[y >= gamma], pdf_0[y >= gamma], color=color_hypothesis[0], alpha=0.1, label='$P_{fa}$')
    ax.fill_between(y[y <= gamma], pdf_1[y <= gamma], color=color_hypothesis[1], alpha=0.1, label='$P_{md}$') 
    
    # Plot Bayes threshold
    ax.vlines(gamma, ymin=0, ymax=pdf_0.max(), lw=2, color='magenta', label='$\gamma$')
    
    # Plot samples
    ax.scatter(y_samples, np.zeros(y_samples.shape), color=color_hypothesis[hypothesis_true], marker='x', label='$y_n$')
    ax.vlines(y_samples.mean(), ymin=0, ymax=pdf_0.max(), lw=2, color=color_hypothesis[hypothesis_test], label=r'$\frac{1}{N}\sum_{n=1}^{N}y_n$')
    
    # Legend
    legend = ax.legend(bbox_to_anchor=[0.5, 1.02], loc='lower center', ncol=4)
    legend.set(title='Légende')

# Widget
slider_kwargs=dict(continuous_update=False, orientation="horizontal")
widget_main = wg.interactive(
    main,
    # number of observations
    N = wg.IntSlider(value=5, min=1, max=10, step=1, description='$N$', **slider_kwargs),
    # costs
    c_fa = wg.FloatLogSlider(value=1, base=10, min=0, max=3, step=1, description=r'$C_{fa}$', **slider_kwargs),
    c_md = wg.FloatLogSlider(value=1, base=10, min=0, max=3, step=1, description=r'$C_{md}$', **slider_kwargs),
    # probability of hypothesis 0
    p_0 = wg.FloatSlider(value=0.5, min=0.1, max=0.9, step=.1, description=r'$p_0$', **slider_kwargs),
    # means hypotheses
    m_0 = wg.FloatSlider(value=-1, min=-5, max=0, step=0.5, description=r'$m_0$', **slider_kwargs),
    m_1 = wg.FloatSlider(value=1, min=0, max=5, step=0.5, description=r'$m_1$', **slider_kwargs),
    # noise variance
    sigma = wg.FloatSlider(value=1, min=1, max=5, step=.5, description=r'$\sigma$', **slider_kwargs),
)

# Output widget
output = widget_main.children[-1]
output.layout=wg.Layout(
    border='1px black solid',
    width="8in", 
    height="4in",
    display='flex',
    flex_flow='column',
    justify_content='space-between',
    align_items='center', 
    align_content='space-between',
)

# Sliders widgets
sliders = widget_main.children[:-1]

# Right panel       
right_panel = wg.Box(
    sliders,
    layout=wg.Layout(
        border='1px black solid',
        height='4in',
        display='flex',
        flex_flow='column',
        justify_content='space-between',
        align_items='center', 
        align_content='space-between',        
    )
)

# Layout
layout = wg.HBox([output, right_panel])
display(layout)
```

### B. Courbes ROC et Tests de Neyman-Pearson

#### 1. Test d'hypothèse simplifié

On considère à présent que $m_0=0$ et $m_1=\rho$. Le test d'hypothèse s'écrit alors:
$$\begin{aligned}
    S
    \ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\
    \frac{\sigma^2 \ln \eta}{N\rho} + \frac{\rho}{2}
\end{aligned}$$

#### 2. Probabilités de détection et fausse alarme

On commence par calculer la probabilité de fausse alarme définie par
$$\begin{aligned}
    P_{fa}
    &=
        \mathbb{P}_{\hat{M}\mid M}(m_1 \mid m_0)
    \\
    &=
        \mathbb{P}_{S\mid M}(S > \gamma \mid m_0)
    \\
    &=
        \int_{\gamma}^{+\infty}
            f_{S\mid M}(s \mid e_0)
        ds
\end{aligned}$$

La variable aléatoire $S$ est la somme de variables **iid** gaussienne selon $\mathcal{N}(0, \sigma^2)$. On a donc $S$ qui est aussi une variable gaussienne de moyenne
$$\begin{aligned}
    \mathbb{E}_{S\mid M}[S\mid m_i] = \frac{1}{N}\sum_{n=1}^{N}\mathbb{E}_{Y_n\mid M}[Y_n\mid m_i] = m_0
\end{aligned}$$
et variance
$$\begin{aligned}
    \mathbb{V}_{S\mid M}[S\mid m_i] = \frac{1}{N^2}\sum_{n=1}^{N}\mathbb{V}_{Y_n\mid M}[Y_n\mid m_i] = \frac{\sigma^2}{N}.
\end{aligned}$$
Ainsi
$$\begin{aligned}
    P_{fa}
    &=
        \int_{\gamma}^{+\infty}
            \sqrt{\frac{N}{2\pi\sigma^2}}
            \exp\left(
                - \frac{N(s - m_0)^2}{2\sigma^2}
            \right)
        ds
    \\
    &=
        \int_{\frac{\sqrt{N}(\gamma-m_0)}{\sigma}}^{+\infty}
            \sqrt{\frac{1}{2\pi}}
            \exp\left(
                - \frac{u^2}{2}
            \right)
        ds
    \\
    &=
        Q\left(\frac{\sqrt{N}(\gamma-m_0)}{\sigma}\right).
\end{aligned}$$
De la même façon, on peut calculer la probabilité de détection comme
$$\begin{aligned}
    P_{d}
    &=
        \mathbb{P}_{\hat{M}\mid M}(m_1 \mid m_1)
    \\
    &=
        \mathbb{P}_{S\mid M}(S > \gamma \mid m_1)
    \\
    &=
        \int_{\gamma}^{+\infty}
            f_{S\mid M}(s \mid e_0)
        ds
    \\
    &=
        Q\left(\frac{\sqrt{N}(\gamma-m_1)}{\sigma}\right).
\end{aligned}$$

#### 3. $P_d$ en fonction de $P_{fa}$

Pour exprimer $P_d$ en fonction de $P_{fa}$, on va utiliser le fait que
$$\begin{aligned}
    P_{fa} 
    &= 
        Q\left(\frac{\sqrt{N}(\gamma-m_0)}{\sigma}\right)
    \\
    Q^{-1}(P_{fa})\frac{\sigma}{\sqrt{N}} + m_0
    &= 
        \gamma
\end{aligned}$$
En remplaçant $\gamma$ dans l'expression de $P_d$, on obtient
la relation cherchée
$$\begin{aligned}
    P_d
    &=
        Q\left(
            \frac{\sqrt{N}}{\sigma}
            \left(Q^{-1}(P_{fa}) \frac{\sigma}{\sqrt{N}} + m_0 - m_1\right)
        \right)
    \\
    &=
        Q\left(
            Q^{-1}(P_{fa})
            -           
            \frac{\sqrt{N}}{\sigma}
            \rho
        \right)
    \\
    &=
        Q\left(
            Q^{-1}(P_{fa})
            -           
            d
        \right)
    &\text{où}&&
    d = \frac{\sqrt{N}}{\sigma} \rho
\end{aligned}$$

#### 4. Courbes ROC

```python
import numpy as np
from matplotlib import pyplot as plt

def roc(d):
    p_fa = np.linspace(0, 1, 100)
    p_d = stats.norm.sf(
        stats.norm.isf(
            p_fa
        )
        -
        d
    )

    fig, ax = plt.subplots()
    ax.set(
        xlabel='$P_{fa}$',
        ylabel='$P_{d}$',
    )
    ax.grid(which='both', axis='both')
    ax.plot(p_fa, p_d, label="$(P_{fa}, P_d)$ décision logique")
    ax.plot(1-p_fa, 1-p_d, label="$(1 - P_{fa}, 1 - P_d)$  décision contraire")
    ax.legend()
# Widget
slider_kwargs=dict(continuous_update=False, orientation="horizontal")
widget_main = wg.interactive(
    roc,
    # number of observations
    d = wg.FloatSlider(value=0, min=0, max=10, step=0.1, description='$d$', **slider_kwargs)
)

# Output widget
output = widget_main.children[-1]
output.layout=wg.Layout(
    border='1px black solid',
    width="8in", 
    height="4in",
    display='flex',
    flex_flow='column',
    justify_content='space-between',
    align_items='center', 
    align_content='space-between',
)

# Sliders widgets
sliders = widget_main.children[:-1]

# Right panel       
right_panel = wg.Box(
    sliders,
    layout=wg.Layout(
        border='1px black solid',
        height='4in',
        display='flex',
        flex_flow='column',
        justify_content='space-between',
        align_items='center', 
        align_content='space-between',        
    )
)

# Layout
layout = wg.HBox([output, right_panel])
display(layout)
```
