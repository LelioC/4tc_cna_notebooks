# Théorie de la Détection
## Communication Numérique Avancées
### Test de Variance

#### A. Détecteur Bayésien

##### 1. Densité de probabilité

Le vecteur $\bm{Y}=[Y[1],\cdots,Y[N]]$ est constitué de composantes **iid** selon
$$\begin{aligned}
    \forall n\in [1,N],\
    \begin{cases}
        Y[n] \sim \mathcal{N}(0, \sigma_0^2)\\
        Y[n] \sim \mathcal{N}(0, \sigma_1^2)
    \end{cases}
\end{aligned}$$
ce qui permet de factoriser la densité de probabilité comme suit
$$\begin{aligned}
    f_{\bm{Y}\mid\Sigma}(\bm{y}\mid\sigma_i)
    &=
        \prod_{n=1}^{N}
            f_{y[n]\mid\Sigma}(y[n]\mid\sigma_i)
    \\
    &=
        \prod_{n=1}^{N}
            \frac{1}{\sqrt{2 \pi \sigma_i^2}}
                \exp\left(
                    -\frac{
                        y[n]^2
                    }{
                        2\sigma_i^2
                    }
                \right)
    \\
    &=
        \frac{1}{(2 \pi \sigma_i^2)^{N/2}}
        \exp\left(
            -\frac{
                \|\bm{y}\|^2
            }{
                2\sigma_i^2
            }
        \right)
    \\
    &=
        \frac{1}{(\pi \lambda_i(\sigma_i))^{N/2}}
        \exp\left(
            -\frac{
                \|\bm{y}\|^2
            }{
                \lambda_i(\sigma_i)
            }
        \right)
    &\text{où on pose}&&
    \lambda_i(\sigma_i) = 2\sigma_i^2
\end{aligned}$$

##### 2. Rapport de vraisemblance

On calcule le rapport de vraisemblance comme suit
$$\begin{aligned}
    \Lambda(\bm{y})
    &=
        \frac{
            f_{\bm{Y}\mid\Sigma}(\bm{y}\mid\sigma_1)
        }{
            f_{\bm{Y}\mid\Sigma}(\bm{y}\mid\sigma_0)
        }
    \\
    &=
        \frac{
            \frac{1}{(\pi \lambda_1(\sigma_1))^{N/2}}
            \exp\left(
                -\frac{
                    \|\bm{y}\|^2
                }{
                    \lambda_1(\sigma_1)
                }
            \right)
        }{
            \frac{1}{(\pi \lambda_0(\sigma_0))^{N/2}}
            \exp\left(
                -\frac{
                    \|\bm{y}\|^2
                }{
                    \lambda_0(\sigma_0)
                }
            \right)
        }
    \\
    &=
        \left(\frac{\lambda_0(\sigma_0)}{\lambda_1(\sigma_1)}\right)^{N/2}
            \exp\left(
                \frac{
                    \|\bm{y}\|^2
                }{
                    \lambda_0(\sigma_0)
                }
                -
                \frac{
                    \|\bm{y}\|^2
                }{
                    \lambda_1(\sigma_1)
                }
            \right)
    \\
    &=
        \left(\frac{\lambda_0(\sigma_0)}{\lambda_1(\sigma_1)}\right)^{N/2}
            \exp\left(
                \|\bm{y}\|^2
                \left(
                    \frac{
                        1       
                    }{
                        \lambda_0(\sigma_0)
                    }
                    -
                    \frac{
                        1
                    }{
                        \lambda_1(\sigma_1)
                    }    
                \right)
            \right)
\end{aligned}$$
> **Note:** On remarque que ce rapport de vraisemblance est identique à celui que l'on obtiendrait en considérant des vraisemblances de lois Exponentielles de paramètres $1/\lambda_i(\sigma_i^2)$.

##### 3. Test optimal de Bayes
Pour prendre une décision sur la valeur de $M$, on va faire un test de Bayes optimal. En utilisant le résultat de la question précédente, le risque de Bayes s'écrit:
$$
\begin{aligned}
        \Lambda(\bm{y})
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
        \eta
    \\
        \left(
            \frac{
                \lambda_0(\sigma_0)
            }{
                \lambda_1(\sigma_1)
            }
        \right)^{N/2}
        \exp\left(
            \|\bm{y}\|^2
            \left(
                \frac{
                    1       
                }{
                   \lambda_0(\sigma_0)
                }
                -
                \frac{
                    1
                }{
                    \lambda_1(\sigma_1)
                }    
            \right)
        \right)
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
        \eta.
\end{aligned}$$
où 
$$\begin{aligned}
    \eta 
    &= 
        \frac{
            C_{\text{fp}}
        }{
            C_{\text{nd}}
        }
        \frac{
            P_M(m_0)
        }{
            P_M(m_1)
        }
\end{aligned}$$
En appliquant la fonction $\ln$ de chaque côté, on obtient
$$
\begin{aligned}
        \frac{N}{2}\ln\frac{\lambda_0(\sigma_0)}{\lambda_1(\sigma_1)}
        +
        \|\bm{y}\|^2
        \left(
            \frac{
                1       
            }{
                \lambda_0(\sigma_0)
            }
            -
            \frac{
                1
            }{
                \lambda_1(\sigma_1)
            }    
        \right)
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
        \ln \eta.
    \\
        \|\bm{y}\|^2
        \left(
            \frac{
                1       
            }{
                \lambda_0(\sigma_0)
            }
            -
            \frac{
                1
            }{
                \lambda_1(\sigma_1)
            }    
        \right)
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
        \ln \eta + \frac{N}{2} \ln\frac{\lambda_1(\sigma_1)}{\lambda_0(\sigma_0)}.
    \\ 
        \|\bm{y}\|^2
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\
        \frac{
            \left(
                \ln \eta 
                + \frac{N}{2} \ln\frac{\lambda_1(\sigma_1)}{\lambda_0(\sigma_0)}
            \right)
        }{
            \frac{1}{\lambda_0(\sigma_0)}
            -
            \frac{1}{\lambda_1(\sigma_1)}
        }
    \\
        \frac{1}{N}
        \sum_{n=1}^{N}
            y[n]^2
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
        \frac{
            2 \ln\eta 
            + N \ln\frac{\lambda_1(\sigma_1)}{\lambda_0(\sigma_0)}
        }{
            2 N
            \left(
                \frac{1}{\lambda_0(\sigma_0)}
                -
                \frac{1}{\lambda_0(\sigma_1)}
            \right)
        }
    \\
        \frac{1}{N}
        \sum_{n=1}^{N}
            y[n]^2
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
            \frac{
                2 \ln\eta 
                + N \ln\frac{2\sigma_1^2}{2\sigma_0^2}
            }{
                2 N
                \left(
                    \frac{1}{2\sigma_1^2}
                    -
                    \frac{1}{2\sigma_0^2}
                \right)
            }
    \\
        \underbrace{
            \frac{1}{N}
            \sum_{n=1}^{N}
                y[n]^2
        }_{S}
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
        \underbrace{
            \frac{
                2 \ln\eta 
                + N \ln\frac{\sigma_1^2}{\sigma_0^2}
            }{
                N
                \left(
                    \frac{1}{\sigma_1^2}
                    -
                    \frac{1}{\sigma_0^2}
                \right)
            }
        }_{\tau}
\end{aligned}$$


##### 4. Détecteur ML et MAP
###### ML

$$\begin{aligned}
    \tau
    &=
        \ln\frac{\sigma_1}{\sigma_0}
        \left(
            \frac{
                1       
            }{
                2\sigma_0^2
            }
            -
            \frac{
                1
            }{
                2\sigma_1^2
            }    
        \right)^{-1}
\end{aligned}$$

###### MAP

$$\begin{aligned}
    \tau
    &=
        \left(
            \frac{1}{N}
            \ln\frac{
                P_{\Sigma}(\sigma_0)
            }{
                P_{\Sigma}(\sigma_1)
            } 
            - \ln\frac{\sigma_0}{\sigma_1}
        \right)
        \left(
            \frac{
                1       
            }{
                2\sigma_1^2
            }
            -
            \frac{
                1
            }{
                2\sigma_0^2
            }    
        \right)^{-1}
\end{aligned}$$

```python
%matplotlib inline
import numpy as np
import ipywidgets as wg
from cna import detection
from scipy import stats
from matplotlib import pyplot as plt

#Main
def main(N, p_0, sigma_0, sigma_1, c_fa, c_md):
    # Prior probabilities for both hypothesis
    p = np.array([p_0, 1 - p_0])

    # Noise mean for both hypothesis
    lambdas = 2*np.array([sigma_0, sigma_1])**2

    # Generate samples of Y
    hypothesis_true = detection.select_hypothesis(p)
    y_samples = detection.generate_samples_exponential(
        lamb=lambdas[hypothesis_true],
        n_samples=N
    )

    # Test
    eta = detection.bayes_threshold(p, c_fa, c_md)
    gamma = detection.test_threshold_exponential(eta, lambdas, N)
    hypothesis_test = detection.bayes_test(y_samples**2, gamma)

    # Figure
    fig, ax = plt.subplots(figsize=[8,3])
    pdf_max = 1 / lambdas.min()
    ax.set(
        xlim=[1e-4, 1e4],
        ylim=[-0.05, pdf_max  + 0.05],
        xscale='log',
        xlabel=r'$y_n^2$ (échelle log.)',
        ylabel=r'$f_{Y^2\mid\Sigma}(y_n^2\mid\sigma_i)$',
    )
    
    # axes 
    y = np.linspace(-1, 10000, 100000)
    pdf_0 = stats.expon(scale=lambdas[0]).pdf(y)
    pdf_1 = stats.expon(scale=lambdas[1]).pdf(y)
    
    # Plot means
    color_hypothesis = ['b', 'g']
    ax.vlines(1/lambdas[0], ymin=0, ymax=stats.expon(scale=lambdas[0]).pdf(1/lambdas[0]), lw=1, ls='--', color=color_hypothesis[0], alpha=0.5 )
    ax.vlines(1/lambdas[1], ymin=0, ymax=stats.expon(scale=lambdas[1]).pdf(1/lambdas[1]), lw=1, ls='--', color=color_hypothesis[1], alpha=0.5 )
    
    # Plot pdfs
    ax.plot(y, pdf_0, color=color_hypothesis[0], lw=1, label=r'$\mathcal{E}(2\sigma_0^2)$' )
    ax.plot(y, pdf_1, color=color_hypothesis[1], lw=1, label=r'$\mathcal{E}(2\sigma_1^2)$' )
    
    # Plot error probabilities
    ax.fill_between(y[y >= gamma], pdf_0[y >= gamma], color=color_hypothesis[0], alpha=0.1, label='$P_{fa}$' )
    ax.fill_between(y[y <= gamma], pdf_1[y <= gamma], color=color_hypothesis[1], alpha=0.1, label='$P_{md}$' ) 
    
    # Plot Bayes threshold
    ax.vlines(gamma, ymin=0, ymax=1/lambdas.min(), lw=2, color='magenta', label='$\gamma$' )
    
    # Plot samples
    ax.scatter(y_samples**2, np.zeros(y_samples.shape), color=color_hypothesis[hypothesis_true], marker='x', label='$y_n^2$' )

    # Plot sample mean
    ax.vlines((y_samples**2).mean(), ymin=0, ymax=1/lambdas.min(), lw=2, color=color_hypothesis[hypothesis_test], label=r'$\frac{1}{N}\sum_{n=1}^{N}y_n^2$' )
    
    # Legend
    legend = ax.legend( bbox_to_anchor=[0.5, 1.02], loc='lower center', ncol=4 )
    legend.set(title='Légende')

# Widget
slider_kwargs=dict(continuous_update=False, orientation="horizontal")
widget_main = wg.interactive(
    main,
    # number of observations
    N = wg.IntSlider(value=5, min=1, max=10, step=1, description='$N$', **slider_kwargs),
    # costs
    c_fa = wg.FloatLogSlider(value=1, base=10, min=0, max=3, step=1, description=r'$C_{fa}$', **slider_kwargs),
    c_md = wg.FloatLogSlider(value=1, base=10, min=0, max=3, step=1, description=r'$C_{md}$', **slider_kwargs),
    # probability of hypothesis 0
    p_0 = wg.FloatSlider(value=0.5, min=0.1, max=0.9, step=.1, description=r'$p_0$', **slider_kwargs),
    # means hypotheses
    sigma_0 = wg.FloatSlider(value=1, min=0.1, max=5, step=0.1, description=r'$\sigma_0$', **slider_kwargs),
    sigma_1 = wg.FloatSlider(value=2, min=0.1, max=5, step=0.1, description=r'$\sigma_1$', **slider_kwargs)
)

# Output widget
output = widget_main.children[-1]
output.layout=wg.Layout(
    border='1px black solid',
    width="75%", 
    height="5in",
    display='flex',
    flex_flow='column',
    justify_content='space-between',
    align_items='center', 
    align_content='space-between',
)

# Sliders widgets
sliders = widget_main.children[:-1]

# Right panel       
right_panel = wg.Box(
    sliders,
    layout=wg.Layout(
        border='1px black solid',
        width='25%',
        height='5in',
        display='flex',
        flex_flow='column',
        justify_content='space-between',
        align_items='center', 
        align_content='space-between',        
    )
)

# Layout
layout = wg.HBox([output, right_panel])
display(layout)
```
