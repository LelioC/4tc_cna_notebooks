### Test Poisson

#### A. Détecteur Bayésien

##### 1. Rapport de vraisemblance

On calcule le rapport de vraisemblance comme étant le ratio des probabilités d'observer le bit $y$ sachant que l'on a transmis un $1$ ou un $0$.
$$\begin{aligned}
    \Lambda(y)
    &=
        \frac{
            P_{Y\mid X}(y\mid 1)
        }{
            P_{Y\mid X}(y\mid 0)
        }
    \\
    &=
        \frac{
            (1 - \lambda_1)^{y}
            \lambda_1^{1 - y}
        }{
            (1 - \lambda_0)^{1 - y}
            \lambda_0^{y}
        }
\end{aligned}$$
Si on transmettait $N$ fois le même bit de façon **iid**, on aurait alors
$$\begin{aligned}
    P_{\bm{Y}\mid X}(\bm{y}\mid x)
    &=
        \prod_{n=1}^{N}
            P_{Y_n\mid X}(y_n\mid x)
    \\
    &=
        \begin{cases}
            \prod_{n=1}^{N}
                (1 - \lambda_1)^{y_n}
                \lambda_1^{1 - y_n}
            \quad\text{if }x=0
            \\
            \prod_{n=1}^{N}
                (1 - \lambda_0)^{1 - y_n}
                \lambda_0^{y_n}
            \quad\text{if }x=1
        \end{cases}
    \\
    &=
        \begin{cases}
            (1 - \lambda_1)^{\sum_{n=1}^{N}y_n}
            \lambda_1^{N - \sum_{n=1}^{N}y_n}
            \quad\text{if }x=0
            \\
            (1 - \lambda_0)^{N - \sum_{n=1}^{N}y_n}
            \lambda_0^{\sum_{n=1}^{N}y_n}
            \quad\text{if }x=1
        \end{cases}            
\end{aligned}$$
et le rapport des vraisemblances s'écrirait alors
$$\begin{aligned}
    \Lambda(y)
    &=
        \frac{
            P_{\bm{Y}\mid X}(\bm{y}\mid 1)
        }{
            P_{\bm{Y}\mid X}(\bm{y}\mid 0)
        }
    \\
    &=
        \frac{
            (1 - \lambda_1)^{\sum_{n=1}^{N}y_n}
            \lambda_1^{N - \sum_{n=1}^{N}y_n}
        }{
            (1 - \lambda_0)^{N - \sum_{n=1}^{N}y_n}
            \lambda_0^{\sum_{n=1}^{N}y_n}
        }
\end{aligned}$$

##### 2. Règle de Bayes

Pour prendre une décision sur la valeur de $X$, on va faire un test de Bayes optimal. En utilisant le résultat de la question précédente, le risque de Bayes s'écrit:
$$\begin{aligned}
    \Lambda(y)
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
    \eta
    \\
    \frac{
        (1 - \lambda_1)^{\sum_{n=1}^{N}y_n}
        \lambda_1^{N - \sum_{n=1}^{N}y_n}
    }{
        (1 - \lambda_0)^{N - \sum_{n=1}^{N}y_n}
        \lambda_0^{\sum_{n=1}^{N}y_n}
    }
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
    \eta.
    \\
    \left.
    \begin{matrix}
        \text{si } y=0&
        \frac{
            \lambda_1
        }{
            1 - \lambda_0
        }\\
        \text{si } y=1&
        \frac{
            1 - \lambda_1
        }{
            \lambda_0
        }
    \end{matrix}
    \right\}    
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
    \eta
    &\text{si}&&
    N=1
\end{aligned}$$
où 
$$\begin{aligned}
    \eta 
    &= 
        \frac{
            C_{\text{fp}}
        }{
            C_{\text{nd}}
        }
        \frac{
            P_M(m_0)
        }{
            P_M(m_1)
        }
    \tag{9}
\end{aligned}$$
En appliquant la fonction $\ln$ de chaque côté de l'avant dernière ligne du calcul précédent, on obtient
$$\begin{aligned}
        \sum_{n=1}^{N} y_n \ln \frac{1 - \lambda_1}{\lambda_0}
        + 
        (N - \sum_{n=1}^{N} y_n) \ln \frac{\lambda_1}{1-\lambda_0}
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
        \ln\eta
    \\
        \sum_{n=1}^{N} y_n
        \left(
            \ln \frac{1 - \lambda_1}{\lambda_0}
            + \ln \frac{1 - \lambda_0}{\lambda_1}
        \right)
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
        \ln\eta + N\ln \frac{1-\lambda_0}{\lambda_1}
    \\
        \sum_{n=1}^{N}
            y_n
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
        \frac{
            \ln\eta
            + N\ln \frac{1 - \lambda_0}{\lambda_1}
        }{
            \ln \frac{1 - \lambda_1}{\lambda_0}
            + \ln \frac{1 - \lambda_0}{\lambda_1}
        }
    \\
        \frac{1}{N}
        \sum_{n=1}^{N}
                y_n
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
        \frac{
            \ln\eta
            + N\ln \frac{1 - \lambda_0}{\lambda_1}
        }{
            N(
                \ln \frac{1 - \lambda_1}{\lambda_0}
                + \ln \frac{1 - \lambda_0}{\lambda_1}
            )
        }
    \\
        \frac{1}{N}
        \sum_{n=1}^{N}
            y_n
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
        \frac{
            \ln \frac{1 - \lambda_0}{\lambda_1}
        }{
            \ln \frac{1 - \lambda_1}{\lambda_0}
            + \ln \frac{1 - \lambda_0}{\lambda_1}
        }
    &\text{(en supposant l'équiprobabilité a priori des bits)}
\end{aligned}$$

##### 3. Risque de Bayes

Dans le cas où $N=1$, $C_{00} = C_{11} = 0$ et $C_{01} = C_{10} = 1$, le risque de Bayes s'écrit
$$\begin{aligned}
    Q(\mathfrak{R})
    &=
        P_X(0) P_{\hat{X}\mid X}(1\mid 0)
        + P_X(1) P_{\hat{X}\mid X}(0\mid 1)
    \\
    &=
        P_X(0) P_{S\mid X}(S > \gamma \mid 0)
        + P_X(1) P_{S\mid X}(S < \gamma \mid 1)
\end{aligned}$$
Comme $\gamma\in[0,1]$ et $S\in[0,1]$, on a:
* $S > \gamma \Rightarrow \sum_{n=1}^{N} y_n > \gamma N$
* $S < \gamma \Rightarrow \sum_{n=1}^{N} y_n < \gamma N$
En posant $S' = \sum_{n=1}^{N} y_n$, où on rappelle que les $y_n$ sont des variables aléatoires iid de Bernoulli, on en déduit que $S'$ est une variable aléatoire suivant la loi Binomiale
* $\mathcal{B}(N; \lambda_0)$ sous l'hypothèse $0$;
* $\mathcal{B}(N; 1 - \lambda_1)$ sous l'hypothèse $1$.

Par conséquent, on a
$$\begin{aligned}
    Q(\mathfrak{R})
    &=
       P_X(0) P_{S'\mid X}(S' > N\gamma \mid 0)
       + P_X(1) P_{S\mid X}(S' < N\gamma \mid 1)
   \\
    &=
       P_X(0) \left(1 - \sum_{s'=0}^{\lfloor N\gamma\rfloor} P_{S'\mid X}(s'\mid 0)\right)
       + P_X(1) \sum_{s'=0}^{\lfloor N\gamma\rfloor} P_{S'\mid X}(s'\mid 0))
    \\
    &=
       P_X(0) + (P_X(1) - P_X(0)) \sum_{s'=0}^{\lfloor N\gamma\rfloor} P_{S'\mid X}(s'\mid 0)
    \\
    &=
       P_X(0) + (P_X(1) - P_X(0)) \sum_{s'=0}^{\lfloor N\gamma\rfloor} \binom{N}{s'} \lambda_0^{s'}(1 - \lambda_0)^{1 - s'}
\end{aligned}$$

##### 4. Règle de Bayes

On suppose $\lambda_0 = \lambda_1 = \lambda$. La règle de Bayes devient
$$\begin{aligned}
    \frac{1}{N}
    \sum_{n=1}^{N}
        y_n
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
    \frac{
            \ln \frac{1-\lambda}{\lambda}
        }{
            \ln \frac{1 - \lambda}{\lambda}
            + \ln \frac{1 - \lambda}{\lambda}
        }
    \\
    \frac{1}{N}
        \sum_{n=1}^{N}
            y_n
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
    \frac{
        1
    }{
        2
    }
\end{aligned}$$

#### (Bonus) Simulation

```python
%matplotlib inline
import numpy as np
import ipywidgets as wg
from cna import detection
from scipy import stats
from matplotlib import pyplot as plt

#Main
def main(N, p_0, lambda_0, lambda_1, c_fa, c_md):    
    # Prior probabilities for both hypothesis
    p = np.array([p_0, 1 - p_0])

    # Noise mean for both hypothesis
    lambdas = np.array([lambda_0, 1 - lambda_1])

    # Generate samples of Y
    hypothesis_true = detection.select_hypothesis(p)
    y_samples = detection.generate_samples_binary(
        lamb=lambdas[hypothesis_true],
        n_samples=N
    )

    # Test
    eta = detection.bayes_threshold(p, c_fa, c_md)
    gamma = detection.test_threshold_binary(eta, lambdas, N)
    hypothesis_test = detection.bayes_test(y_samples, gamma)

    # Figure
    fig, ax = plt.subplots(figsize=[8,3])
    pdf_max = 1
    ax.set(
        xlim=[-0.5, 1.5],
        ylim=[-0.05, pdf_max  + 0.05],
        xlabel=r'$y_n$',
        ylabel=r'$f_{Y\mid X}(y_n\mid x)$',
    )
    
    # axes 
    y = np.array([0, 1])
    pdf_0 = stats.bernoulli(p=lambdas[0]).pmf(y)
    pdf_1 = stats.bernoulli(p=lambdas[1]).pmf(y)
    
    # Plot means
    color_hypothesis = ['b', 'g']
    ax.vlines(stats.bernoulli(p=lambdas[0]).mean(), ymin=0, ymax=1, lw=1, ls='--', color=color_hypothesis[0], alpha=0.5 )
    ax.vlines(stats.bernoulli(p=lambdas[1]).mean(), ymin=0, ymax=1, lw=1, ls='--', color=color_hypothesis[1], alpha=0.5 )
    
    # Plot pdfs
    ax.plot(y, pdf_0, color=color_hypothesis[0], ls='none', lw=1, marker='o', label=r'$\mathcal{B}(\lambda_0)$' )
    ax.plot(y, pdf_1, color=color_hypothesis[1], ls='none', lw=1, marker='o', label=r'$\mathcal{B}(1-\lambda_1)$' )
    
    # Plot error probabilities
    ax.fill_between(y[y >= gamma], pdf_0[y >= gamma], color=color_hypothesis[0], alpha=0.1, label='$P_{fa}$' )
    ax.fill_between(y[y <= gamma], pdf_1[y <= gamma], color=color_hypothesis[1], alpha=0.1, label='$P_{md}$' ) 
    
    # Plot Bayes threshold
    ax.vlines(gamma, ymin=0, ymax=1/lambdas.min(), lw=2, color='magenta', label='$\gamma$' )
    
    # Plot samples
    ax.scatter(y_samples**2, np.zeros(y_samples.shape), color=color_hypothesis[hypothesis_true], marker='x', label='$y_n$' )

    # Plot sample mean
    ax.vlines((y_samples**2).mean(), ymin=0, ymax=1/lambdas.min(), lw=2, color=color_hypothesis[hypothesis_test], label=r'$\frac{1}{N}\sum_{n=1}^{N}y_n$' )
    
    # Legend
    legend = ax.legend( bbox_to_anchor=[0.5, 1.02], loc='lower center', ncol=4 )
    legend.set(title='Légende')

# Widget
slider_kwargs=dict(continuous_update=False, orientation="horizontal")
widget_main = wg.interactive(
    main,
    # number of observations
    N = wg.IntSlider(value=5, min=1, max=10, step=1, description='$N$', **slider_kwargs),
    # costs
    c_fa = wg.FloatLogSlider(value=1, base=10, min=0, max=3, step=1, description=r'$C_{fa}$', **slider_kwargs),
    c_md = wg.FloatLogSlider(value=1, base=10, min=0, max=3, step=1, description=r'$C_{md}$', **slider_kwargs),
    # probability of hypothesis 0
    p_0 = wg.FloatSlider(value=0.5, min=0.1, max=0.9, step=.1, description=r'$p_0$', **slider_kwargs),
    # means hypotheses
    lambda_0 = wg.FloatSlider(value=0.5, min=0.1, max=0.9, step=0.1, description=r'$\sigma_0$', **slider_kwargs),
    lambda_1 = wg.FloatSlider(value=0.5, min=0.1, max=0.9, step=0.1, description=r'$\sigma_1$', **slider_kwargs)
)

# Output widget
output = widget_main.children[-1]
output.layout=wg.Layout(
    border='1px black solid',
    width="75%", 
    height="5in",
    display='flex',
    flex_flow='column',
    justify_content='space-between',
    align_items='center', 
    align_content='space-between',
)

# Sliders widgets
sliders = widget_main.children[:-1]

# Right panel       
right_panel = wg.Box(
    sliders,
    layout=wg.Layout(
        border='1px black solid',
        width='25%',
        height='5in',
        display='flex',
        flex_flow='column',
        justify_content='space-between',
        align_items='center', 
        align_content='space-between',        
    )
)

# Layout
layout = wg.HBox([output, right_panel])
display(layout)
```
