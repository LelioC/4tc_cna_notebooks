# Théorie de la Détection
## Communication Numérique Avancées
### Test Poisson

#### A. Détecteur Bayésien

##### 1. Domaines de définition


###### Domaine des événements
Le domaine des événements est **discret**: $\mathcal{E}=\{\lambda_0, \lambda_1\}$.

###### Domaine des observations
Le domaine des observations est aussi **discret**: $\mathcal{Y}=\N$.

##### 2. Probabilités

Du fait du caractère **iid** des mesures, on peut factoriser la probabilité conditionnelle sur les observations du nombre de paquets par trames:
$$\begin{aligned}
    P_{\bm{Y}\mid \Lambda}(\bm{y}\mid\lambda_i)
    &=
        \prod_{n=1}^{N}
            P_{Y[n]\mid \Lambda}(y[n]\mid\lambda_i)
    \\
    &=
        \prod_{n=1}^{N}
            \frac{\lambda_i^{y[n]}}{y[n]!}
            \exp(-\lambda_i)
    \\
    &=
        \exp(-N\lambda_i)
        \prod_{n=1}^{N}
            \frac{\lambda_i^{y[n]}}{y[n]!}
\end{aligned}$$

#### 3. Rapport de vraisemblance


$$\begin{aligned}
    V(\bm{y})
    &=
        \frac{
            P_{\bm{Y}\mid \Lambda}(\bm{y}\mid\lambda_1)
        }{
            P_{\bm{Y}\mid \Lambda}(\bm{y}\mid\lambda_0)
        }
    \\
    &=
        \frac{
            \exp(-N\lambda_1)
            \prod_{n=1}^{N}
                \frac{\lambda_1^{y[n]}}{y[n]!}
            }{
            \exp(-N\lambda_0)
                \prod_{n=1}^{N}
                    \frac{\lambda_0^{y[n]}}{y[n]!}
                }
    \\
    &=
        \exp(N(\lambda_0 - \lambda_1))
        \prod_{n=1}^{N}
            \left(
                \frac{
                    \lambda_1
                }{
                    \lambda_0
                }
            \right)^{y[n]}
    \\
    &=
        \exp(N(\lambda_0 - \lambda_1))
        \left(
                \frac{
                    \lambda_1
                }{
                    \lambda_0
                }
            \right)^{\sum_{n=1}^{N} y[n]}
\end{aligned}$$

#### 4. Détection de base optimale

Pour prendre une décision sur la valeur de $M$, on va faire un test de Bayes optimal. En utilisant le résultat de la question précédente, le risque de Bayes s'écrit:
$$\begin{aligned}
    \Lambda(\bm{y})
    \ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
    \eta
    \\
    \exp(N(\lambda_0 - \lambda_1))
    \left(
        \frac{
            \lambda_1
        }{
            \lambda_0
        }
    \right)^{\sum_{n=1}^{N} y[n]}
    \ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
    \eta.
\end{aligned}$$
où 
$$\begin{aligned}
    \eta 
    &= 
        \frac{
            C_{\text{fp}}
        }{
            C_{\text{nd}}
        }
        \frac{
            P_M(m_0)
        }{
            P_M(m_1)
        }
\end{aligned}$$
En appliquant la fonction $\ln$ de chaque côté, on obtient
$$\begin{aligned}
    N(\lambda_0 - \lambda_1)
    + 
        (\ln \lambda_1 - \ln \lambda_0)
        \sum_{n=1}^{N} y[n]
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
    \ln\eta.
    \\
    (\ln \lambda_1 - \ln \lambda_0)
    \sum_{n=1}^{N} y[n]
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
    \ln\eta + N (\lambda_1 - \lambda_0).
    \\
    \underbrace{
        \frac{1}{N}
        \sum_{n=1}^{N} y[n]
    }_{S}
    &\ \underset{\mathbb{H}_1}{\overset{\mathbb{H}_0}{\lessgtr}}\ 
    \underbrace{
        \frac{
            \ln\eta + N (\lambda_1 - \lambda_0).
        }{
            N(\ln \lambda_1 - \ln \lambda_0)
        }
    }_{\gamma}
\end{aligned}$$

##### 4. Probabilités d'erreur

###### Fausse alarme

$$\begin{aligned}
    p_{fa}
    &= 
        P_{S\Lambda}(S > \gamma, \Lambda=\lambda_0)
    \\
    &=
        P_\Lambda(\lambda_0)
        P_{S\Lambda}(S > \gamma \mid \Lambda=\lambda_0) 
    \\
    &=
        P_\Lambda(\lambda_0)
        \sum_{s>\lceil\gamma\rceil}^{+\infty}
            P_{S\mid\Lambda}(S = s\mid\Lambda=\lambda_0) 
\end{aligned}$$
En utilisant le fait qu'une somme de $N$  aléatoires **iid** selon une loi de Poisson de paramètre $\lambda_0$ est aussi une variable aléatoire de Poisson de paramètre $N\lambda_0$, on obtient
$$\begin{aligned}
    p_{fa}
    &=
        P_\Lambda(\lambda_0)
        \sum_{s>\lceil\gamma\rceil}^{+\infty}
            \frac{(N\lambda_0)^s}{s!}\exp(-N\lambda_0)
\end{aligned}$$

###### Non-détection

$$\begin{aligned}
    p_{nd}
    &=
        P_{S\Lambda}(S < \gamma, \Lambda=\lambda_1)
    \\
    &=
        P_\Lambda(\lambda_1)
        P_{S\Lambda}(S < \gamma \mid \Lambda=\lambda_1) 
    \\
    &=
        P_\Lambda(\lambda_1)
        \sum_{s=0}^{\lfloor\gamma\rfloor}
            P_{S\mid\Lambda=\lambda_1}(S = s\mid\Lambda=\lambda_1) 
\end{aligned}$$
En utilisant le fait qu'une somme de $N$  aléatoires **iid** selon une loi de Poisson de paramètre $\lambda_1$ est aussi une variable aléatoire de Poisson de paramètre $N\lambda_1$, on obtient
$$\begin{aligned}
    p_{nd}
    &=
        P_\Lambda(\lambda_1)
        \sum_{s=0}^{\lfloor\gamma\rfloor}
            \frac{(N\lambda_1)^s}{s!}\exp(-N\lambda_1)
\end{aligned}$$


```python
%matplotlib inline
import numpy as np
import ipywidgets as wg
from cna import detection
from scipy import stats
from matplotlib import pyplot as plt

#Main
def main(N, p_0, lambda_0, lambda_1, c_fa, c_md):
    # Prior probabilities for both hypothesis
    p = np.array([p_0, 1 - p_0])

    # Noise mean for both hypothesis
    lambdas = np.array([lambda_0, lambda_1])

    # Generate samples of Y
    hypothesis_true = detection.select_hypothesis(p)
    y_samples = detection.generate_samples_poisson(
        lamb=lambdas[hypothesis_true], 
        n_samples=N
    )

    # Test
    eta = detection.bayes_threshold(p, c_fa, c_md)
    gamma = detection.test_threshold_poisson(eta, lambdas, N)
    hypothesis_test = detection.bayes_test(y_samples, gamma)

    # Figure
    fig, ax = plt.subplots(figsize=[8,3])
    pdf_max = stats.poisson(lambdas.min()).pmf(lambdas.min())
    ax.set(
        xlim=[0, 10],
        ylim=[-0.05, pdf_max  + 0.05],
    )
    
    # axes 
    y = np.arange(0, 100)
    pdf_0 = stats.poisson(lambdas[0]).pmf(y)
    pdf_1 = stats.poisson(lambdas[1]).pmf(y)
    
    # Plot means
    color_hypothesis = ['b', 'g']
    ax.vlines( lambdas[0], ymin=0, ymax=stats.poisson(lambdas[0]).pmf(lambdas[0]), lw=1, ls='--', color=color_hypothesis[0], alpha=0.5 )
    ax.vlines( lambdas[1], ymin=0, ymax=stats.poisson(lambdas[1]).pmf(lambdas[1]), lw=1, ls='--', color=color_hypothesis[1], alpha=0.5 )
    
    # Plot pdfs
    ax.plot(y, pdf_0, color=color_hypothesis[0], lw=1, label=r'$\mathcal{P}(\lambda_0)$' )
    ax.plot(y, pdf_1, color=color_hypothesis[1], lw=1, label=r'$\mathcal{P}(\lambda_1)$' )
    
    # Plot error probabilities
    ax.fill_between(y[y >= gamma], pdf_0[y >= gamma], color=color_hypothesis[0], alpha=0.1, label='$P_{fa}$' )
    ax.fill_between(y[y <= gamma], pdf_1[y <= gamma], color=color_hypothesis[1], alpha=0.1, label='$P_{md}$' ) 
    
    # Plot Bayes threshold
    ax.vlines(gamma, ymin=0, ymax=pdf_0.max(), lw=2, color='magenta', label='$\gamma$' )
    
    # Plot samples
    ax.scatter(y_samples, np.zeros(y_samples.shape), color=color_hypothesis[hypothesis_true], marker='x', label='$y_n$' )

    # Plot sample mean
    ax.vlines(y_samples.mean(), ymin=0, ymax=pdf_0.max(), lw=2, color=color_hypothesis[hypothesis_test], label=r'$\frac{1}{N}\sum_{n=1}^{N}y_n$' )
    
    # Legend
    legend = ax.legend( bbox_to_anchor=[0.5, 1.02], loc='lower center', ncol=4 )
    legend.set(title='Légende')

# Widget
slider_kwargs=dict(continuous_update=False, orientation="horizontal")
widget_main = wg.interactive(
    main,
    # number of observations
    N = wg.IntSlider(value=5, min=1, max=10, step=1, description='$N$', **slider_kwargs),
    # costs
    c_fa = wg.FloatLogSlider(value=1, base=10, min=0, max=3, step=1, description=r'$C_{fa}$', **slider_kwargs),
    c_md = wg.FloatLogSlider(value=1, base=10, min=0, max=3, step=1, description=r'$C_{md}$', **slider_kwargs),
    # probability of hypothesis 0
    p_0 = wg.FloatSlider(value=0.5, min=0.1, max=0.9, step=.1, description=r'$p_0$', **slider_kwargs),
    # means hypotheses
    lambda_0 = wg.IntSlider(value=1, min=1, max=5, step=1, description=r'$\lambda_0$', **slider_kwargs),
    lambda_1 = wg.IntSlider(value=2, min=1, max=5, step=1, description=r'$\lambda_1$', **slider_kwargs)
)

# Output widget
output = widget_main.children[-1]
output.layout=wg.Layout(
    border='1px black solid',
    width="8in", 
    height="4in",
    display='flex',
    flex_flow='column',
    justify_content='space-between',
    align_items='center', 
    align_content='space-between',
)

# Sliders widgets
sliders = widget_main.children[:-1]

# Right panel       
right_panel = wg.Box(
    sliders,
    layout=wg.Layout(
        border='1px black solid',
        height='4in',
        display='flex',
        flex_flow='column',
        justify_content='space-between',
        align_items='center', 
        align_content='space-between',        
    )
)

# Layout
layout = wg.HBox([output, right_panel])
display(layout)
```
